```python
superr = sample.drop(['Postal Code'], axis=1)
superr = superr.drop(['Country'], axis=1)
```

```python
superr.describe()
```

```python
corrl = superr.corr()
corrl
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(corrl, annot = True)
plt.show()
```

```python
superr['Sub-Category'].value_counts()
```

```python
plt.figure(figsize=(16,8))
plt.bar('Sub-Category', 'Category', data=superr)
plt.show()
```

```python
print (superr['Ship Mode'].value_counts())
plt.figure(figsize=(10,6))
sns.countplot(x=superr['Ship Mode'])
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x=superr['Segment'])
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.countplot(x=superr['City'], order=(superr['City'].value_counts().head(50)).index)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x=superr['Sub-Category'])
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.pie(superr['Category'].value_counts(), labels=superr['Category'].value_counts().index, startangle=180, radius=1)
plt.title('Main category of operation')
plt.show()
print (superr['Category'].value_counts())
```

```python
plt.figure(figsize=(12,6))
plt.pie(superr['Sub-Category'].value_counts(), labels=superr['Sub-Category'].value_counts().index, startangle=180, radius=1)
plt.show()
print (superr['Sub-Category'].value_counts())
```
